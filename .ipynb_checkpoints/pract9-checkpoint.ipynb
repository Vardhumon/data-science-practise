```python
import seaborn as sns
import pandas as pd
```

```python
# Load the titanic dataset
df = sns.load_dataset("titanic")
```

```python
# Prepare the data for boxplot
p = sns.catplot(
    x="sex",
    y="age",
    hue="survived",
    col="survived",  # Separate plots for survived/not survived
    kind="box",
    showmeans=True,  # Show mean as a point
    data=df,
)
```

```python
# Rotate x labels to prevent overlapping
p.despine(left=True)
p.set_xlabels(rotation=45)
plt.show()
```
