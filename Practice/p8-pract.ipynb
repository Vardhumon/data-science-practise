2. Write a code to check how the price of the ticket (column name:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

```python
df
```

```python
df.info()
```

```python
plt.figure(figsize=(5,4))
sns.histplot(df,x='fare',bins=20)
plt.title("Fare Distribution")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(data=df,x='sex',palette='pastel')
plt.title("Passenger Count By SEX")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='class', data=df, palette='pastel', order=['First', 'Second', 'Third'])
plt.title("Passenger Count By CLASS")
plt.show()
```
