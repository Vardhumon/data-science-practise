```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Student_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Math_Score': [85, 90, 78, 92, 88, np.nan, 75, 82, 95, 98],
    'Science_Score': [88, 92, 80, 94, 90, 85, 72, 79, 93, 96],
    'English_Score': [82, 88, 75, 90, 85, 80, 70, 78, 92, 94]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
nums=['Math_Score','Science_Score','English_Score']
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(data=df[nums])
plt.title('Boxplot of Numeric Variables')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
df['Math_Score_Log']=np.log(df['Math_Score'])
df['Science_Score_Log']=np.log(df['Science_Score'])
df['English_Score_Log']=np.log(df['English_Score'])
```
